```python
import ast
```

```python
mov.head()
```

```python
def convert(obj):
    l=[]
    for i in ast.literal_eval(obj):
        l.append(i["name"])
    return l
```

```python
mov["genres"]=mov["genres"].apply(convert)
mov["keywords"]=mov["keywords"].apply(convert)
# mov["cast"]=mov["cast"].apply(convert)
mov.head()
```

```python
mov["cast"]=mov["cast"].apply(convert)
mov.head()
```

```python
def convert2(obj):
    l=[]
    for i in ast.literal_eval(obj):
        if i["job"]=="Director" or i["job"]=="Original Music Composer":
            l.append(i["name"])
    return l
```

```python
mov["crew"]=mov["crew"].apply(convert2)
mov.head()
```

```python
mov["overview"]=mov["overview"].apply(lambda x:x.split())
```

```python
mov.head()
```

```python
mov["genres"]=mov["genres"].apply(lambda x:[i.replace(" ","") for i in x])
mov["cast"]=mov["cast"].apply(lambda x:[i.replace(" ","") for i in x])
mov["keywords"]=mov["keywords"].apply(lambda x:[i.replace(" ","") for i in x])
mov["crew"]=mov["crew"].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
mov.head()
```

```python
mov["tags"]=mov["overview"]+mov["genres"]+mov["keywords"]+mov["cast"]+mov["crew"]
mov.head()
```

```python
new_mov=mov[["movie_id","title","tags"]]
new_mov
```

```python
import pandas as pd
import nltk
from nltk.corpus import stopwords

# Download the stop words data (you only need to do this once)
nltk.download('stopwords')
def remove_stopwords(text):
    stop_words = set(stopwords.words('english'))
    words = nltk.word_tokenize(text)
    return ' '.join([word for word in words if word.lower() not in stop_words])
new_mov['tags'] = new_mov['tags'].apply(remove_stopwords)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
cosine_similarity(temp)
```

```python
cosine_similarity(temp).shape
```

```python
sim=cosine_similarity(temp)
sim[0]
```

```python


def recommend(movie,n):
    movie_id=new_mov[new_mov["title"]==movie].index[0]
    distances=sim[movie_id]
    movies_list=sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:n+1]
    movies = [new_mov.iloc[score[0]]['title'] for score in movies_list]
    print(movies)
```

```python
recommend("Batman v Superman: Dawn of Justice",5)
```

```python
recommend("Interstellar",10)
```

```python
recommend("The Wolf of Wall Street",5)
```

```python
recommend("Spectre",7)
```
